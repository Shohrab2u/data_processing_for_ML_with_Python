## Rename the column names appropriately

```python
print(df.columns)
```

```python
df = df.rename(columns={'Mat':'Match', 
                        'Inns':'Innings',
                        'Wkts': 'Wickets',
                        'Ave': 'Average',
                        'Econ': 'Economy rate',
                        'SR': 'Strike Rate',
                        4: 'Four wickets',
                        5: 'Five wickets',})

display(df.head())
```

## How many players played for ICC?

```python
# splitting the 'Player' column to get the information about 'Country'
df[["Player_Name", "Country"]] = df['Player'].str.split("(", expand=True)

display(df.head())

# dropping the 'Player' columns
df = df.drop('Player', axis=1)

# remove the ")" from the 'Country' column
df['Country'] = df['Country'].str.replace(")", "")

display(df.head())

# rearrange the columns
new_col_sequence = ['Player_Name', 'Country', 'Span', 'Match', 'Innings', 'Balls', 'Runs', 'Wickets',
       'Average', 'Economy rate', 'Strike Rate', 'Four wickets', 'Five wickets',]


df = df[new_col_sequence]

display(df.head())
```

## How many different countries are present in this dataset?

```python
df['Country'].value_counts()
```

## Which player(s) had played for the longest period of time?

```python
# Defining two new column 'start year' and 'end yar'

df[['start_year', 'end_year']] = df['Span'].str.split("-", expand=True)

# Dropping the 'Span' column from the dataset
df = df.drop("Span", axis=1)

display(df.head())
```

```python
# Converting the data type of 'start_year' and 'end_year' column
df['start_year'] = df['start_year'].astype('int')
df['end_year'] = df['end_year'].astype('int')

# Creating new column 'years_played'
df['years_played'] = (df['end_year'] - df['start_year']) + 1

# Dropping the column 'start_year' and 'end_year'
df = df.drop(['start_year', 'end_year'], axis=1)
display(df.head())
```

## Which player(s) had played for the shortest period of time?

```python
df.sort_values(by='years_played', ascending=False).head(10)

```
